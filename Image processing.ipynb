Importing important libraries

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as signal
import scipy.ndimage as ndimage
```

```python
Noise = cv2.imread('/content/noisy_4012980830_5d2b84158d_c.jpg')
plt.imshow(Noise)
plt.show()
```

Preprocesing image and changing to grayscale

```python
image = cv2.imread('Noise.jpg')
gray_image = cv2.cvtColor(Noise, cv2.COLOR_BGR2GRAY)
resized_image = cv2.resize(gray_image, (600, 500))
normalized_image = resized_image / 255.0


plt.imshow(normalized_image, cmap='gray')
plt.title('Preprocessed Image')
plt.show()
```

Applying median filter

```python
median_filtered = cv2.medianBlur(resized_image, 3)  #best kernel(3)

plt.imshow(median_filtered, cmap='gray')
plt.title('Median Filtered Image')
plt.show()
```

Applying mean filter

```python
mean_filtered = ndimage.uniform_filter(normalized_image, size=5)

plt.imshow(mean_filtered, cmap='gray')
plt.title('Mean Filtered Image')
plt.show()
```

**Comparison and Discussion**<br>

Median Filter: Excellent for removing salt-and-pepper noise because it replaces each pixel with the median value of its neighbors. This effectively eliminates extreme outlier values (the salt and pepper).

Mean Filter: Tends to blur edges and details more than a median filter. It averages the pixel values, which can smooth out noise but also reduce sharpness.

For salt-and-pepper noise, the median filter generally performs better. This is because the median is less sensitive to extreme outliers that characterize this type of noise. The mean filter, on the other hand, can be influenced by these outliers, resulting in less effective noise removal.

Visualizing the pictures

```python
plt.figure(figsize=(25, 25))

plt.subplot(131)
plt.imshow(normalized_image, cmap='gray')
plt.title('Original Image')

plt.subplot(132)
plt.imshow(median_filtered, cmap='gray')
plt.title('Median Filtered')

plt.subplot(133)
plt.imshow(mean_filtered, cmap='gray')
plt.title('Mean Filtered')

plt.tight_layout()
plt.show()
```
